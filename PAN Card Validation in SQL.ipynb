```python
%sql
CREATE OR REPLACE FUNCTION fn_check_adjacent_characters(p_str STRING)
RETURNS BOOLEAN
LANGUAGE PYTHON
AS $$
if p_str is None or len(p_str) < 2:
    return False
for i in range(len(p_str) - 1):
    if p_str[i] == p_str[i+1]:
        return True
return False
$$;

CREATE OR REPLACE FUNCTION fn_check_sequence_characters(p_str STRING)
RETURNS BOOLEAN
LANGUAGE PYTHON
AS $$
if p_str is None or len(p_str) < 2:
    return False
#It will check first pair and id squence of characters not found it will exit loop and return False. 
for i in range(len(p_str) - 1):
    if (ord(p_str[i+1]) - ord(p_str[i])) != 1:
        return False
return True
$$;
```

```python
#Logic to check At least one single sequential pair
for i in range(len(p_str) - 1):
    if (ord(p_str[i+1]) - ord(p_str[i])) == 1:
        return True
return False
```
